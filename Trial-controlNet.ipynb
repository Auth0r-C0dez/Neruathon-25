```python
import torch
from diffusers import StableDiffusionPipeline

# Load Stable Diffusion v1.5 model
sd_model = "runwayml/stable-diffusion-v1-5"
pipeline = StableDiffusionPipeline.from_pretrained(sd_model, torch_dtype=torch.float16)
pipeline.to("cuda")
```

```python
from diffusers import ControlNetModel

# Load the ControlNet model for edge detection
controlnet_model = "lllyasviel/control_v11p_sd15_canny"
controlnet = ControlNetModel.from_pretrained(controlnet_model, torch_dtype=torch.float16)
controlnet.to("cuda")
```

```python
from diffusers import StableDiffusionControlNetPipeline, UniPCMultistepScheduler

# Initialize Stable Diffusion with ControlNet
pipe = StableDiffusionControlNetPipeline.from_pretrained(
    sd_model, controlnet=controlnet, torch_dtype=torch.float16
)

# Use UniPC Multistep Scheduler for faster, high-quality inference
pipe.scheduler = UniPCMultistepScheduler.from_config(pipe.scheduler.config)

# Enable memory optimizations
pipe.enable_model_cpu_offload()
```

```python
import os
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm  # Progress bar

# Ensure edge directory exists
os.makedirs(edge_dir, exist_ok=True)

# Supported image formats
valid_extensions = (".jpg", ".jpeg", ".png", ".bmp", ".tiff")

# Process images and apply edge detection
for file_name in tqdm(os.listdir(image_dir), desc="Processing Images"):
    input_path = os.path.join(image_dir, file_name)
    output_path = os.path.join(edge_dir, file_name)

    # Skip non-image files
    if not file_name.lower().endswith(valid_extensions):
        print(f"⚠️ Skipping non-image file: {file_name}")
        continue

    try:
        # Load image in grayscale
        image = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)

        if image is None:
            print(f"❌ Error loading image: {file_name}, skipping...")
            continue

        # Apply Canny edge detection
        edges = cv2.Canny(image, 100, 200)  # Adjust thresholds if needed

        # Convert to RGB (for consistency)
        edges = cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB)

        # Convert to PIL format and save
        edge_image = Image.fromarray(edges)

        # Ensure output format is PNG to avoid format issues
        if not output_path.lower().endswith(".png"):
            output_path = output_path.rsplit(".", 1)[0] + ".png"

        edge_image.save(output_path)

    except Exception as e:
        print(f"❌ Error processing {file_name}: {e}")

print(f"✅ Edge-detected images saved in: {edge_dir}")
```

```python
import torch
from PIL import Image
from diffusers import StableDiffusionControlNetPipeline, ControlNetModel
import matplotlib.pyplot as plt

# Load the Stable Diffusion and ControlNet model
sd_model = "runwayml/stable-diffusion-v1-5"

controlnet_model = "lllyasviel/control_v11p_sd15_canny"

# Load the pre-trained models
controlnet = ControlNetModel.from_pretrained(controlnet_model, torch_dtype=torch.float16).to("cuda")

pipeline = StableDiffusionControlNetPipeline.from_pretrained(
    sd_model, controlnet=controlnet, torch_dtype=torch.float16
).to("cuda")

# Load the edge-detected image (replace with actual image path)
edge_image_path = "/content/floor_plan_dataset/edges/floor_001.png"  # Update this path
edge_image = Image.open(edge_image_path).convert("RGB")

# Generate the floor plan using Stable Diffusion + ControlNet
prompt = "A professional, detailed floor plan, modern architectural design, high resolution"
negative_prompt = "low quality, blurry, unrealistic"

output = pipeline(
    prompt=prompt,
    negative_prompt=negative_prompt,
    image=edge_image,
    num_inference_steps=50,  # Increase for better quality
    guidance_scale=7.5,  # Adjust for stronger ControlNet influence
).images[0]

# Save and display the generated image
output.save("/content/generated_floorplan.png")

plt.figure(figsize=(10, 10))
plt.imshow(output)
plt.axis("off")
plt.show()
```
