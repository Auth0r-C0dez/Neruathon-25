```python
import torch
from diffusers import StableDiffusionXLImg2ImgPipeline
from PIL import Image
import os
import pandas as pd
from tqdm import tqdm  # For progress bar

# Load the SDXL Turbo pipeline
pipeline = StableDiffusionXLImg2ImgPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-turbo",
    torch_dtype=torch.float16
).to("cuda")

# Define prompts for different room types
room_prompts = {
    "kitchen": "A high-resolution top-down view of a modern kitchen. Features cabinets, stove, sink, refrigerator, and a small dining table. Clean and well-lit.",
    "bedroom": "A high-resolution top-down view of a cozy bedroom. Includes a bed, nightstand, wardrobe, lamp, and a soft rug. Minimalist and elegant design.",
    "balcony": "A high-resolution top-down view of a spacious balcony. Has potted plants, a small table with chairs, and a scenic view beyond the railing.",
    "corridor": "A high-resolution top-down view of a corridor. Well-lit with wooden flooring, doors on both sides, and decorative elements on the walls."
}

# Output directories
output_dir = "generated_images"
os.makedirs(output_dir, exist_ok=True)

# CSV file to store image paths
csv_file = "generated_images.csv"

# Number of images per category
num_images_per_category = 200

# Data list to store image info
data = []

# Generate and save images
for room_type, prompt in room_prompts.items():
    room_dir = os.path.join(output_dir, room_type)
    os.makedirs(room_dir, exist_ok=True)

    print(f"Generating {num_images_per_category} images for {room_type}...")

    for i in tqdm(range(num_images_per_category)):
        image = pipeline(prompt=prompt, num_inference_steps=4).images[0]  # Fast generation
        image_path = os.path.join(room_dir, f"{room_type}_{i+1}.png")
        image.save(image_path)
        data.append([room_type, image_path])

# Save paths to CSV
df = pd.DataFrame(data, columns=["Room Type", "Image Path"])
df.to_csv(csv_file, index=False)

print(f"All images saved in '{output_dir}', and paths stored in '{csv_file}'")
```
